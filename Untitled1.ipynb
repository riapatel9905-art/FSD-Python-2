Pandas series

```python
import pandas as pd
a=[1,7,2]
v=pd.Series(a)
print(v)
print(type(v))
```

```python
print(v[1])
```

names of index

```python
a=[1,7,2]
v=pd.Series(a,index=["x","y","z"])
print(v)
print(type(v))
```

```python
a=(1.4,2,3,4)
v=pd.Series(a)
print(v)
```

```python
s_n=["ram","shyam","radha","geeta","seeta"]
marks=[50,60,70,80,90]
x=pd.Series(marks,index=s_n)
print(x)
```

```python
s_n=["ram","shyam","radha","geeta","seeta"]
marks=[50,60,70,80,90]
x=pd.Series(marks,index=s_n,name="student result")
print(x)
```

```python
calories={'day1':400,'day2':380,'day3':430,'day4':390}
m_v=pd.Series(calories,index=['day1','day2'])
print(m_v)
```

```python
calories={'day1':400,'day2':380,'day3':430,'day4':390}
m_v=pd.Series(calories,name="calories burn per day")
print(m_v)
```
