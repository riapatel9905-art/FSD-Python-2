## ch-4 Regression-Model Training and evaluation


--> Linear Regression is a statistical method used to modal the relationship between a dependent variable and one or more indeoendent variables
--> The goal of LR is to find the best fitting line that describes the relationship between the variables
--> Dependent Variable(y)-Predict
--> Independent Variable(x) - Predictors or variables we use to predict the dependent variables

MAE - mean absolute error
MSE - mean square error
R^2 - 1-SSres/SStotal     (SSres):Sum of square : residual             (SStotal):Sum of square : total

```python
import pandas as pd
```

```python
df =pd.read_csv("eg1.csv")
df.head()
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
print(x)
print(y)
```

```python
print(type(x))
print(type(y))
```

```python
print(x.shape)
print(y.shape)
```

```python
#split the data in training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1) #if the value of random state
                                                                                #is changed then the values will be different for testing
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#creating linearRegression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y_pred)
```

```python
diff=pd.DataFrame({"actual":y,"Predicted":y_pred})
diff
```

```python
print("Coefficient",lr.coef_)
print("Intercept",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y,y_pred))
print("MSE: ",metrics.mean_squared_error(y,y_pred))
print("R2 Score : ",metrics.r2_score(y,y_pred))
```

```python
y_pred=lr.predict([[30]]) #whenever we enter a value manually it should be 2d
print(y_pred)
```

```python
#after the model is trained
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
diff=pd.DataFrame({"actual":y_test,"Predicted":y_pred})
diff
```

```python
df1=pd.read_csv("student_scores.csv")
df1
```

```python
#Predicting student Scores Based on Study hours
x1=df1.iloc[:,0:1]
y1=df1.iloc[:,-1]
print(x1)
print(y1)
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.1,random_state=1) #if the value of random state
                                                                                #is changed then the values will be different for testing
print(x1_train.shape)
print(x1_test.shape)
print(y1_train.shape)
print(y1_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x1,y1)
y_pred1=lr.predict(x1)
print(y_pred1)
```

```python
diff1=pd.DataFrame({"actual":y1,"Predicted":y_pred1})
diff1
```

```python
print("Coefficient",lr.coef_)
print("Intercept",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y1,y_pred1))
print("MSE: ",metrics.mean_squared_error(y1,y_pred1))
print("R2 Score : ",metrics.r2_score(y1,y_pred1))
```

```python
import matplotlib.pyplot as plt
#print(f"MSE: {mse:.2f}")
#print(f"R^2 score: {r2:2f}")
plt.scatter(x1,y1,color="blue",label="Actual Score")
plt.plot(x1,y_pred1,color="red",label="predicted line")
plt.xlabel("Hours studied")
plt.ylabel("Exam Score")
plt.title("Linear Regression:Hours vs score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
error=[]
for i in range(1,43):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    error.append(metrics.mean_absolute_error(y_test,y_pred))
print(error)
```

```python
#Dict
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
error_dict={}
for i in range(1,43):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    error_dict[i]=metrics.mean_absolute_error(y_test,y_pred)
print(error_dict)
```
