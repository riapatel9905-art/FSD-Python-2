# Data Frame


Pandas Dataframe is a 2-dimensional data structure like a 2 dimensional array or a table with rows and columns

```python
data={'calories':[400,420,380,390],"duration":[50,40,45,60]}
df=pd.DataFrame(data)
df
```

```python
df.loc[0]
```

```python
df.iloc[0]
```

```python
df=pd.DataFrame(data,index=['day1','day2','day3','day4'])
```

```python
df.loc['day1']
```

```python
df.iloc['day1']
```

```python
df[::2]
```

```python
data={'Name':['Alice','Bob','Charlie'],'Age':[25,30,35],'city':['New York','Los angeles','Chicago']}
df=pd.DataFrame(data)
df
```

```python
df.loc[0:2]
```

```python
df.iloc[0:2]
```

```python
data={'A':[1,2,3,4],'B':[5,6,7,8]}
df=pd.DataFrame(data,index=['row1','row2','row3','row4'])
df
```

```python
df.loc['row':'row4','B']=99
df
```

```python
df.iloc[1:3,0]=0
df
```

```python
df['C']=[10,20,30,40]
df
```

```python
df.loc[df['A']==0,'new_col']='zero'
df
```

```python
df.loc['row5']=[10,20,30,40]
df
```

# Set_index()

Used to set one or more columns of a dataframe as the index

```python
data={'Name':['Alice','Bob','Charlie'],'Age':[25,30,35],'city':['Ny','LA','Chicago']}
df=pd.DataFrame(data)
df
```

```python
df.set_index('Name',drop=False)
```

```python
df
```

```python
df.set_index(['Name','Age'],drop=False,inplace=True)
df
```

# reset_index()
Used to manage and reset the index od a DataFrame

```python
data={'Name':['Alice','Bob','Charlie'],'Age':[25,30,35],'city':['Ny','LA','Chicago']}
df=pd.DataFrame(data,index=['a','b','c'])
df
```

```python
df_reset=df.reset_index()
df_reset
```
