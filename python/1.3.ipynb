## Parallel_coordinates

```python
from pandas.plotting import parallel_coordinates
df=pd.read_csv('auto-mpg.csv')
p=parallel_coordinates(df,'cylinders',cols=['acceleration','mpg','origin'], color=['red','blue','green','yellow','black'])
```
