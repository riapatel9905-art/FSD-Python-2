```python
import pandas as pd
```

```python
#to merge if both the names are different then this method is used
a1=pd.DataFrame({"id":[1,2,3],
                "name":["amit","riya","john"]})
a2=pd.DataFrame({"s_id":[1,2,4],"marks":[85,90,75]})
r=pd.merge(a1,a2,left_on="id",right_on="s_id")
print(r)
```

```python
d1=pd.DataFrame({"Physics":[70,80,90]},index=[101,102,103])
d2=pd.DataFrame({"Name":["Alice","bob","charlie"]},index=[101,102,103])
r=pd.merge(d1,d2,left_index=True,right_index=True)
print(r)
```

```python
a1=pd.DataFrame({"id":[1,2,3],
                "name":["amit","riya","john"]})
a2=pd.DataFrame({"Score":[90,80,70]},index=[1,2,3])
r=pd.merge(a1,a2,left_on="id",right_index=True)
print(r)
```

###  groupby()

```python
data={'product category':['Electronics','Furniture','Electronics','Furniture'],'sales':[200,150,300,100]}
df=pd.DataFrame(data)
df
```

```python
r=df.groupby("product category")['sales'].sum()
print(r)
```

```python
r=df.groupby("product category")['sales'].mean()
print(r)
```

```python
r=df.groupby("product category")['sales'].median()
print(r)
```

```python
r=df.groupby("product category")['sales'].describe()
print(r)
```

```python
r=df.groupby("product category")['sales'].sample()
print(r)
```

```python
r=df.groupby("product category")['sales'].size()
print(r)
```

```python
r=df.groupby("product category")['sales'].groups
print(r)
```

```python
r=df.groupby("product category")['sales'].nunique()
print(r)
```

```python
r=df.groupby("product category")['sales'].nth(0)
print(r)
```

```python
r=df.groupby("product category")['sales'].nth(1)
print(r)
```

```python
r=df.groupby("product category",axis=0,as_index=True,sort=True)['sales'].count()
print(r)
```

```python
r=df.groupby("product category",axis=0,as_index=False,sort=True)['sales'].count()
print(r)
```

```python
df=pd.DataFrame({'Region':['east','east','west','west','east','west'],
                "sales":[100,150,200,120,110,130],
                "manager":['a','b','c','d','e','f']})
df
```

```python
def sales_range(group):
    return group['sales'].max()-group['sales'].min() #max-min
sales=df.groupby("Region").apply(sales_range)
print(sales)
```

```python
df=pd.DataFrame({"Dept":['sales','sales','HR','HR','IT','IT'],
                "Sales":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
r=df.groupby('Dept').agg('mean')
print(r)
```

```python
df=pd.DataFrame({"Dept":['sales','sales','HR','HR','IT','IT'],
                "Sales":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
r=df.groupby('Dept').agg(['mean','min','max'])
print(r)
```

```python
df=pd.DataFrame({"Dept":['sales','sales','HR','HR','IT','IT'],
                "Salary":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
r=df.groupby('Dept').agg({'Salary':['mean','max'],"Bonus":'sum'})
print(r)
```

```python
df=pd.DataFrame({"Dept":['sales','sales','HR','HR','IT','IT'],
                "Salary":[50000,60000,40000,45000,70000,75000],
                "Bonus":[5000,6000,2000,2500,8000,8500]})
r=df.groupby('Dept').agg(MeanSalary=("Salary","mean"),MaxSalary=('Salary','max'),TotalBonus=('Bonus','sum'))
print(r)
```
