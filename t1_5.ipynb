# Drop and remove duplicates

```python
import pandas as pd
```

```python
data={'A':["TeamA","TeamB","TeamB","TeamC","TeamA"],
      'B':[50,40,40,30,50],
      'C':[True,False,False,False,True]}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.duplicated() #compares the entire rows
```

```python
df.duplicated(keep='first') #checks from the top
```

```python
df.duplicated(keep='last') #checks from the bottom
```

```python
df.duplicated(keep=False) #gives true for all the duplicates
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(keep="first")
```

```python
df.drop_duplicates(keep="last")
```

```python
df.drop_duplicates(keep=False)
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

### result_type()

```python
df=pd.DataFrame({'width':[1,2],'height':[3,4]})
def get_stats(row):
    return[row['width']* row['height'],2*(row['width']+row['height'])]
df
```

```python
df.apply(get_stats,axis=1,result_type='expand')
```

```python
df.apply(get_stats,axis=1,result_type='broadcast')
```

```python
df.apply(get_stats,axis=1,result_type='reduce')
```

### args and kwargs parameters

```python
def add_values(x,y,z):
    return x + y + z
df=pd.DataFrame({'A':[1,2],'B':[10,20]})
df=df.apply(add_values,args=(1,2))
print(df)
```

```python
#kwargs = Keyword arguements
def multiply_by_m(x,m):
    return x*m
df=pd.DataFrame({'A':[1,2],'B':[10,20]})
df=df.apply(multiply_by_m,m=10)
print(df)
```

### replace() function

```python
df1=pd.DataFrame({"marks":[80,85,0,90],
                  "Attendance":[75,80,0,85],
                  "grade":["A","B","F","A"]})
print(df1)
result=df1.replace([80,75],[90,80])
print(result)
```

```python
df1=pd.DataFrame({"marks":[80,85,0,90],
                  "Attendance":[75,80,0,85],
                  "grade":["A","B","F","A"]})
print(df1)
result=df1.replace({80:75,90:80})
print(result)
```

```python
df1["marks"]=df1["marks"].replace(0,50)
df1
```

```python
df1["Attendance"]=df1["Attendance"].replace(0,60)
df1
```

### concate()

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'C':[7,8]})
result=pd.concat([df1,df2],axis=0)
result
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'C':[7,8]})
result=pd.concat([df1,df2],axis=1)
result
```

```python
pd.concat([df1,df2],join="inner") #only keeps overlap
```

```python
pd.concat([df1,df2],join="inner",ignore_index=True) #resets the index to 0,1,2,....
```

```python
#sorts columns/index if they don't align
df1=pd.DataFrame({'E':[1,2],'B':[3,4]})
df2=pd.DataFrame({'C':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],sort=False,axis=1)
result
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'C':[7,8]})
result=pd.concat([df1,df2],sort=True)
result
```

### merge() function

```python
df1=pd.DataFrame({
    "employee":["A","B","C"],
    "Department":["IT","HR","Finance"]
})
df2=pd.DataFrame({
    "employee":["A","B","D"],
    "Salary":[50000,40000,30000]
})
result=pd.merge(df1,df2,on="employee",how="inner")
result
```

```python
result=pd.merge(df1,df2,on="employee",how="outer")
result
```
