```python
import pandas as pd
df=pd.read_csv("Advertising.csv")
df.head()
               
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
x=df[["TV","Radio","Newspaper"]]
y=df["Sales"]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) 
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x.columns
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
diff=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
diff
```

```python
print("Coefficient: ",lr.coef_)
print("Intercept",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score : ",metrics.r2_score(y_test,y_pred))
```

# POLYLINEAR REGRESSION

```python
df=pd.read_csv("polylinearregression.csv")
df
```

```python
x=df[['Temperature']]
y=df['Pressure']
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
print(x_poly)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_poly,y)
y_pred=lr.predict(x_poly)
```

```python
from sklearn.metrics import r2_score
print("R2 score : ",r2_score(y,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
plt.scatter(x,y,color="blue")
plt.plot(x,y_pred,color='red')
plt.show()
```

# Polynomial Regression

```python
df=pd.read_csv('car_data.csv')
df
```

```python
df.columns
```

```python
x=df[['Horsepower', 'Weight']]
y=df['MPG']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
print(x_poly)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_poly,y,test_size=0.2,random_state=1)  # random_state is used for getting accurate value of train,test

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print("R2 score : ",r2_score(y_test,y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
import pandas as pd

x=[1,2,3,4,5]
y=[1,8,27,64,125]

x=pd.DataFrame(x)
y=pd.DataFrame(y)

poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)

model=LinearRegression()
model.fit(x_poly,y)
y_pred=model.predict(x_poly)

print("R2 Score : ",r2_score(y,y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
import pandas as pd

x=[1,2,3,4,5]
y=[1,8,27,64,125]

x=pd.DataFrame(x)
y=pd.DataFrame(y)

poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)

model=LinearRegression()
model.fit(x_poly,y)
y_pred=model.predict(x_poly)

print("R2 Score : ",r2_score(y,y_pred))
```
