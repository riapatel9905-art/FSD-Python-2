```python
import pandas as pd
```

```python
df=pd.read_csv("book1.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
y=df.iloc[:,0:1]
x=df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) 
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(y_train,x_train)
x_pred=lr.predict(y_test)
print(x_pred)
```

```python
diff=pd.DataFrame({"Actual":x_test,"Predicted":x_pred})
diff
```

```python
print("Coefficient: ",lr.coef_)
print("Intercept",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(x_test,x_pred))
print("MSE: ",metrics.mean_squared_error(x_test,x_pred))
print("R2 Score : ",metrics.r2_score(x_test,x_pred))
```

```python
import matplotlib.pyplot as plt
#plt.scatter(df["cgpa"],df["package"])
plt.scatter(y_test,x_test)
plt.plot(y_test,x_pred,color="red")
plt.show()
```

# Multiple Linear Regression

```python
df=pd.read_csv("car data.csv")
df
```

```python
print(df["Car_Name"].nunique())
```

```python
df.info()
```

```python
df.drop("Car_Name",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Age"]=2026-df["Year"]
```

```python
df.columns
```

```python
df.head()
```

```python
df.drop("Year",axis=1,inplace=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(data=df,drop_first=True)
df.head()
```

```python
df.shape
```

```python
y=df["Selling_Price"]                      #x=df.iloc[:,1:]
                                           #y=df.iloc[:,0]
x=df.drop("Selling_Price",axis=1)
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) 
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
diff=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
diff
```

```python
print("Coefficient: ",lr.coef_)
print("Intercept",lr.intercept_)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score : ",metrics.r2_score(y_test,y_pred))
```
